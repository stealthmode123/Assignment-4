```python
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/gender-prediction.csv')
New_df = df
```

```python
New_df = New_df.drop(['beard','hair_length','scarf','eye_color','gender'], axis=1)
New_df
```

```python
labels = LabelEncoder()
beard_encoded = labels.fit_transform(df['beard'])
hair_length_encoded = labels.fit_transform(df['hair_length'])
scarf_encoded = labels.fit_transform(df['scarf'])
eye_color_encoded = labels.fit_transform(df['eye_color'])
gender_encoded = labels.fit_transform(df['gender'])

New_df['beard'] = beard_encoded
New_df['hair_length'] = hair_length_encoded
New_df['scarf'] = scarf_encoded
New_df['eye_color'] = eye_color_encoded
New_df['gender'] = gender_encoded
New_df
```

```python
x = New_df.drop('gender', axis=1)
y = New_df['gender']
```

```python
from sklearn.model_selection import ShuffleSplit,cross_val_score

model = DecisionTreeClassifier()

shuffle_split=ShuffleSplit(test_size=0.2,train_size=0.8,n_splits=10)

score=cross_val_score(model,x,y,scoring='f1',cv=shuffle_split)
mean_score = cross_val_score(model, x, y, scoring='f1', cv = shuffle_split).mean()
std_score = cross_val_score(model, x, y, scoring='f1', cv = shuffle_split).std()

print("Cross Validation F1 Scores: ", score,'\n')
print("Average F1 Score: ", mean_score,'\n')
print("Standard Deviation F1 Score: ", std_score)
```

30% data was used for testing, 50% percent was used for trainning and value of splits was n=10

```python
from sklearn.model_selection import LeavePOut,cross_val_score

model = DecisionTreeClassifier()

lpo = LeavePOut(p=2)
lpo.get_n_splits(x)

score=cross_val_score(model,x,y,scoring='f1',cv=lpo)
mean_score = cross_val_score(model, x, y, scoring='f1', cv = lpo).mean()
std_score = cross_val_score(model, x, y, scoring='f1', cv = lpo).std()

print("Cross Validation F1 Scores: ", score,'\n')
print("Average F1 Score: ", mean_score,'\n')
print("Standard Deviation F1 Score: ", std_score)
```

Value of p = 1 so one was used for testing rest for trainning.
